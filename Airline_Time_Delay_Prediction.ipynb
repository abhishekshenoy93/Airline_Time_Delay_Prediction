# Process categoty with OneHotEncode
## Pipeline for processing data

```python
# transform catogry into numeric value
x = original_airline_data.drop(['Delays','DelaysN','Year','FlightNum'],axis=1)
y = original_airline_data['Delays']

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

cat_transformer = Pipeline(steps=[("onehot", OneHotEncoder(handle_unknown='ignore'))])
cat_attr = ['Month','DayofMonth','DayOfWeek','UniqueCarrier','Origin','Dest']
preprocessor = ColumnTransformer(transformers=[("cat", cat_transformer, cat_attr)])
```

# Machine Learn -- RF, DT, SVM, LR
## Training Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, KFold, cross_validate, cross_val_score
from sklearn.metrics import accuracy_score
import pickle


base_models = [("dt_clf", DecisionTreeClassifier(random_state=42)),
               ("rf_clf", RandomForestClassifier(n_estimators = 300,random_state=42)),
               ("svm_clf",svm.NuSVC(random_state=42,gamma='auto')),
               ('lr_clf', LogisticRegression(random_state=42))]

kfolds = 10
split = KFold(n_splits=kfolds, shuffle=True, random_state=42)

for name, model in base_models:
    # pack preprocessing of data and the model in a pipeline:
    model_steps = Pipeline(steps=[('preprocessor', preprocessor),
                                 ('model', model)])
    
    # get cross validation score for each model:
    cv_results = cross_val_score(model_steps, 
                                 x, y, 
                                 cv=split,
                                 scoring="accuracy",
                                 n_jobs=-1)
    # Save the model as a pickle in a file 
     
    
    # output:[]
    min_score = round(min(cv_results), 4)
    max_score = round(max(cv_results), 4)
    mean_score = round(np.mean(cv_results), 4)
    std_dev = round(np.std(cv_results), 4)
    print(f"{name} cross validation accuarcy score: {mean_score} +/- {std_dev} (std) min: {min_score}, max: {max_score}")
```

## Training via Random Forest and Check Accuracy Details 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, KFold, cross_validate, cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import joblib



x_train, x_test, y_train, y_test = train_test_split(x,y,test_size= 0.33, random_state = 21)

kfolds = 10
split = KFold(n_splits=kfolds, shuffle=True, random_state=42)

model_steps = Pipeline(steps=[('preprocessor', preprocessor),
                              ('model', RandomForestClassifier(n_estimators = 300,random_state=21))])


# get cross validation score for each model:
y_pred = model_steps.fit(x_train,y_train).predict(x_test)

joblib.dump(model_steps, 'rf_clf.pkl')
print('model saved')
print('\n')
matrix = confusion_matrix(y_test, y_pred)
print(matrix )
print('\n')
score = round(accuracy_score(y_test, y_pred),4)*100
print(f'Accuracy of ths model is {score}')
print('\n')
report = classification_report(y_test, y_pred)
print(report)
```
